```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Function to simulate coin flips
def simulate_coin_flips(n_flips):
    outcomes = [random.choice(['Heads', 'Tails']) for _ in range(n_flips)]
    heads_count = outcomes.count('Heads')
    tails_count = outcomes.count('Tails')
    return heads_count, tails_count

# Function to calculate probabilities
def calculate_probabilities(heads, tails):
    total = heads + tails
    return (heads / total) * 100, (tails / total) * 100

# Running the simulation
n_flips = 1_000_000  # Adjust for desired number of flips
heads, tails = simulate_coin_flips(n_flips)
prob_heads, prob_tails = calculate_probabilities(heads, tails)

# Visualization
labels = ['Heads', 'Tails']
counts = [heads, tails]

plt.figure(figsize=(8, 5))
plt.bar(labels, counts, color=['skyblue', 'salmon'])
plt.title(f"Coin Flip Outcomes: {n_flips} Flips")
plt.ylabel("Counts")
plt.show()

# Confidence Interval Approximation
def confidence_interval(n, prob, z=1.96):
    standard_error = (prob * (1 - prob) / n) ** 0.5
    return prob - z * standard_error, prob + z * standard_error

conf_int_heads = confidence_interval(n_flips, prob_heads / 100)
conf_int_tails = confidence_interval(n_flips, prob_tails / 100)

# Results
print(f"Number of Flips: {n_flips}")
print(f"Heads: {heads} ({prob_heads:.2f}%)")
print(f"Tails: {tails} ({prob_tails:.2f}%)")
print(f"95% CI for Heads: {conf_int_heads}")
print(f"95% CI for Tails: {conf_int_tails}")
```
